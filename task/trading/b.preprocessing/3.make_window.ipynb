# 데이터셋 조정

(Count, Features, Label) 데이터를 (Count, Rows, Cols, Label) 형태로 변경

## 라이브러리 불러오기

```python
import pickle

import numpy as np
import pandas as pd
import tensorflow as tf
```

## 데이터 불러오기

```python
train_df = pd.read_csv("/home/freeman/projects/data/trading/data/train_5930.csv", sep=",")
test_df = pd.read_csv("/home/freeman/projects/data/trading/data/test_5930.csv", sep=",")
data_df = pd.concat([train_df, test_df])
```

```python
data_df.shape
```

## DataFrame을 리스트로 변환

```python
window_size = 52         # date, label 제외한 컬럼의 크기와 같도록(굳이 할 필요는 없음)
loop_count = data_df.shape[0] - window_size
```

```python
data_x = []
data_y = []
dates = []
```

### 윈도우 형태의 데이터 생성

```python
for i in range(loop_count):
    data_x.append(data_df.iloc[i:i+window_size, 1:-1].to_numpy())
    data_y.append(data_df.iloc[i+window_size, -1:].to_numpy())
    dates.append(data_df.iloc[i+window_size, :1].to_numpy())
```

```python
data_x
```

### 레이블을 이용해 변경 값 확인

```python
n = 101
data_y[n], data_df.iloc[n+window_size, -1:], dates[n], data_df.iloc[n+window_size, :1]
```
