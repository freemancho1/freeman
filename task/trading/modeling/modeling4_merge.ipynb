## KOSPI 100

```python
kospi_100_df = pd.read_csv(os.path.join(INDEXES_INFO_PATH, "KOSPI 100 내역.csv"),
                           delimiter=",", encoding="UTF8", skiprows=[0],
                           names=['date', 'close_100', 'open_100', 'high_100', 'low_100', 'volume_100', 'rate_100'])
```

```python
num_columns = kospi_100_df.columns
for column_name in num_columns[1:5]:
    kospi_100_df[column_name] = kospi_100_df[column_name].apply(lambda x: x.replace(",", ""))
kospi_100_df["volume_100"] = kospi_100_df["volume_100"].apply(lambda x: x.replace("K", "").replace("M", ""))
kospi_100_df["rate_100"] = kospi_100_df["rate_100"].apply(lambda x: x.replace("%", ""))
```

```python
kospi_100_df[num_columns[0]] = pd.to_datetime(kospi_100_df[num_columns[0]])
kospi_100_df[num_columns[1:]] = kospi_100_df[num_columns[1:]].apply(pd.to_numeric)
```

```python
kospi_100_df.set_index("date", inplace=True)
kospi_100_df
```
