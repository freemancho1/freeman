# 모델 생성

## SimpleGAN class

### 훈련 데이터 준비

* GAN은 강화학습 개념으로 `레이블`이 필요없고, 시스템에서 일괄적으로 제공하는 `진짜`라는 레이블을 가지고 학습함

```python
from tensorflow.keras.datasets.mnist import load_data

# 데이터 전처리(차원확장, 정규화 등)는 모델 클래스에서 직접 수행함
(train_x, _), (_, _) = load_data()
```

### 모델 훈련

```python
gan_hyper_parameter = {
    "epochs"            : 10000,
    "batch_size"        : 128,
    "latent_z_dim"      : 100,
    "num_display_log"   : 20,
}

try: 
    gan = SimpleGAN(train_x, **gan_hyper_parameter, tf_log_level=2)
    gan.fit()
    gan.display_last_generation_images()
except Exception as e:
    print(e)
```

### 숫자별 모델 훈련

#### 9

```python
from tensorflow.keras.datasets.mnist import load_data
import numpy as np

(train_x, train_y), (_, _) = load_data()
train_x9 = train_x[np.where(train_y == 9)]

gan_hyper_parameter = {
    "epochs"            : 10000,
    "batch_size"        : 128,
    "latent_z_dim"      : 100,
    "num_display_log"   : 20,
}

try: 
    gan = SimpleGAN(train_x9, **gan_hyper_parameter, tf_log_level=2)
    gan.fit()
    gan.display_last_generation_images()
except Exception as e:
    print(e)
```
